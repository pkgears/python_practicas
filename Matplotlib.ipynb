## Representación gráfica de datos

Si a la función de trazados se le da una matriz de datos, la usará como coordenadas en el eje vertical y utilizará el índice de cada punto de datos en el array como la coordenada horizontal

```python
lista = [1,2,3,5,8,7,10,14]

plt.plot(lista)
plt.show()
```

```python
lista2 = [1,4,5,7,8,9,12,14]

plt.plot(lista2, lista)
plt.show()
```

```python
# Pueden modificarse las figuras de los ejes para que la figura no se vea tan ajustada

plt.plot([-3,-1,0,5,7,9],[-1,1,4,6,7,8])
plt.show()
```

```python
plt.plot([-3,-1,0,5,7,9],[-1,1,4,6,7,8])
plt.axis([-4, 10, -3, 10])
plt.show()
```

```python
# Se sigue el mismo procedimiento para pintar un funcion matematica

import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
# Tambien puede modificarse el estilo de la grafica para que tenga mas informacion

plt.plot(x,y)
plt.title("Square function")
plt.xlabel("X")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
# Pueden superponerse graficas y cambiar el estilo de la lineas

import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'r')
plt.grid(True)
plt.show()
```

```python
# Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda

import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
# Tambien pueden crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denomina subplot

import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g')
plt.grid(True)

plt.show()
```

```python
# Para que las graficas no esten tan ajustadas, podemos hacer la figura mas grande

# Tambien pueden crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denomina subplot

import numpy as np

plt.figure(figsize=(12,5))

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.grid(True)
plt.xlabel("x1", fontsize=14)
plt.ylabel("y1=x1**2", fontsize=14)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2=x2**2", fontsize=14)
plt.grid(True)

plt.show()
```

## Scatter plots

```python
from numpy.random import rand
 
x,y = rand(2,100)
```

```python
plt.scatter(x,y, color='green')
plt.show()
```

```python
x,y = rand(2,100)
x2,y2 = rand(2,100)

plt.scatter(x,y, c="red")
plt.scatter(x2, y2)

plt.show()
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(211) # 2 filas, 1 columna, 1er grafico

plt.hist(data, bins=50, rwidth=0.8) # bins = numero de barras, rwidth = ancho de las barras  
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(211)

data2 = [1.5, 1.8, 2.9, 4, 4.3, 4.5, 5, 7]

plt.subplot(212) # 2 filas, 1 columna, 2do grafico

plt.hist(data2, bins=20, rwidth=0.8)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
